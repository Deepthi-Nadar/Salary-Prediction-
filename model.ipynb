```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
db=pd.read_csv("employees.csv")

db
```

```python
db.info()
```

```python
db.drop(columns=["First Name","Last Name","Email",	"Phone",	"Gender", "Age", "Department"])
```

```python
X=db[["Job Title",	"Years Of Experience"]]
```

```python
X
```

```python
y=db["Salary"]
```

```python
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
cf = ["Job Title"]
nf = ["Years Of Experience"]
```

```python
preprocessor = ColumnTransformer([
    ("job", OneHotEncoder(), cf)], remainder="passthrough")
```

```python
preprocessor
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([("preprocessing", preprocessor), ("model", LinearRegression())])
```

```python
print(pipeline)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2)
```

```python
pipeline.fit(train_X, train_y)
```

```python
y_pred = pipeline.predict(test_X)
y_pred
```

```python

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(test_y, y_pred)
mae = mean_absolute_error(test_y, y_pred)
mse = mean_squared_error(test_y, y_pred)
```

```python
print(r2)
print(mae)
print(mse)
```

```python


result = pd.DataFrame({
    "Actual": test_y,
    "Predicted": y_pred
})

print(result.head())
```
